Note: when the prediction residuals have zero mean, the  score is identical to the Explained Variance score.


```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(1000, 1, noise = 20, random_state=42)

if X.shape[1] == 1:
    X_ = X.flatten()
```

```python
sns.scatterplot(x = X_, y = y, alpha = 0.5)
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, lr.predict(X))
```

```python
lr.score(X,y)
```
